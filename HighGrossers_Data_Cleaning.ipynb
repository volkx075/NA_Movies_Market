```python
# Add our dependencies
import csv
import os
import pandas as pd
import numpy as py
```

```python
# Read in CSV file and print as a DataFrame
highestgrossers_original_df = pd.read_csv('Resources/HighestGrossers.csv')
highestgrossers_original_df.head()
```

```python
highestgrossers_original_df.tail()
```

```python
highestgrossers_clean_df = highestgrossers_original_df
```

```python
highestgrossers_clean_df.tail()
```

```python
highestgrossers_clean_df.fillna("Action", inplace=True)
```

```python
highestgrossers_clean_df.tail()
```

```python
highestgrossers_clean_df.head()
```

```python
highestgrossers_clean_df.dtypes
```

```python
col_to_check= ["TICKETS SOLD"]
```

```python
highestgrossers_clean_df[col_to_check] = highestgrossers_clean_df[col_to_check].replace({",": ""}, regex=True)
```

```python
highestgrossers_clean_df[col_to_check] 
```

```python
cols_to_check= ["TOTAL FOR YEAR", "TOTAL IN 2019 DOLLARS"]
```

```python
highestgrossers_clean_df[cols_to_check] = highestgrossers_clean_df[cols_to_check].replace({r'\$': ''}, regex=True)
```

```python
highestgrossers_clean_df[cols_to_check] = highestgrossers_clean_df[cols_to_check].replace({',': ''}, regex=True)
```

```python
highestgrossers_clean_df[cols_to_check] 
```

```python
highestgrossers_clean_df
```

```python
convert_dict = {"TOTAL FOR YEAR": float,
               "TOTAL IN 2019 DOLLARS": float,
               "TICKETS SOLD": float}
```

```python
highestgrossers_clean_df = highestgrossers_clean_df.astype(convert_dict)
```

```python
highestgrossers_clean_df.dtypes
```

```python
# Export to CSV
highestgrossers_clean_df.to_csv("Resources/HighestGrossers_Clean.csv")
```
